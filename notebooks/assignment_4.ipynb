```python
import geopandas as gpd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
pluto = Path('~/Documents/GTECH/Geocomp_731/assignments/inputs/pluto.geojson')
pluto_lic = Path('~/Documents/GTECH/Geocomp_731/assignments/inputs/pluto-LIC.geojson')
```

```python
def df_from_pluto(pluto, to_crs='EPSG:32618'):
    df = gpd.read_file(pluto)
    assert df.crs.srs.lower() == 'epsg:4326'
    df.to_crs(to_crs, inplace=True)
    
    return df
```

```python
df_astoria = df_from_pluto(pluto)
df_lic = df_from_pluto(pluto_lic)
```
